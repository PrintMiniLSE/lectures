# Strings

* Ordered sequences of characters
* Immutable

```python
x = 'my string'

# Capitalises the first character (chr) of the string
x = x.capitalize() # assign it back to x but if you do x.capitalize the original str doesnt change. so by doing x = x.capitalize() we overwrite but do not change it. 

# prints the string
print(x)

# prints the chr at index 3
print(x[3])

# prints the last chr
print(x[-1])

# print a range 
# NOTE: not inclusive of the last index, 4 chrs because python starts at 0 ÎßàÏßÄÎßâ Ïù∏Îç±Ïä§ Ìè¨Ìï®ÏïàÌïòÎäîÍ±∞!!
print(x[0:4])

# Index one to the last index
# Again, not inclusive
print(x[1:-1])

# EXTENDED SLICING 
# Get every other (2) item in the string.
print(x[::2]) #prints the first character, so x[0], x[2], x[4]
print(x[0:5:2]) #ÏõêÏÜå ÏÉàÍ∞ú Ïù¥Í±∞ Îã§Ïãú Ï†ïÎ¶¨Ìï¥ 
#x[start:stop:step]
# Reverse steps, every other chr
print(x[::-2])
print(x[::-1]) #recalls every characters backwards
```
